```python
import spacy
import en_core_web_sm
import geonamescache
import pandas as pd 
```

```python
nlp = en_core_web_sm.load()
```

```python
gc = geonamescache.GeonamesCache()
countries_cache = gc.get_countries_by_names()
countries_list = list(countries_cache.keys())
```

```python
headline = []
with open("headlines.txt") as FileObj:
    for line in FileObj:
        lineStrip = line.rstrip()
        headline.append(lineStrip)
```

```python
disease = [None] * len(headline)
country = [None] * len(headline)
city = [None] * len(headline)
for i,lineStrip in enumerate(headline):
    doc = nlp (lineStrip)
    found_city = False
    found_country = False
    found_disease = False
    possible_diseases = []
    for ent in doc.ents:  
        if ent.text in countries_list and not found_country:
            country[i]= ent.text
            found_country = True
        elif ent.label_ == "GPE" and not found_city : 
            city[i] = ent.text
            found_city = True
        else :
            possible_diseases.append(ent.text)
            found_disease = True
    if not found_city:
        city[i] = "N/A";
    if not found_country:
        country[i] = "N/A"
    if not found_disease:
        disease[i] = "N/A"
    else:
        disease[i] = " ".join(possible_diseases)

```

```python
df = pd.DataFrame(list(zip(headline,city,country)), 
               columns =['headline', 'cities', 'ciybtrues']) 
```
